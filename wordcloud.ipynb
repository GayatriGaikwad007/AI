```python

import matplotlib.pyplot as plt
from wordcloud import WordCloud

# List of anime-related words (feel free to add more)
anime_words = """
Naruto Bleach Dragonball OnePiece AttackOnTitan MyHeroAcademia DemonSlayer FullmetalAlchemist
DeathNote HunterXHunter TokyoGhoul SwordArtOnline FairyTail CowboyBebop OnePunchMan JoJo
NeonGenesisEvangelion Evangelion Pokemon SailorMoon SpiritedAway StudioGhibli Akira SamuraiChamploo
Trigun YuYuHakusho YourName TokyoRevengers MobPsycho GhostInTheShell
"""

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(anime_words)

# Display the word cloud using matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Turn off axis
plt.show()
```
